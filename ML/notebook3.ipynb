```python
import sklearn;

#sklearn.show_versions()

from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

##### 1. Load wine data set

```python
wine = load_wine()
features = wine.data
labels = wine.target

features
```

##### 2. Split de Data (80% om te studeren, 20% voor het examen)

```python
features_train, features_test, labels_train, labels_test = train_test_split(features, labels, test_size=0.2, random_state=42)
```

##### 3. Initialiseer het Model
###### We kijken naar de 3 'nearest neighbors' om een beslissing te nemen

```python
model = KNeighborsClassifier(n_neighbors=3) # je kan voor 3, 5 of 7 kiezen. 3 betekent naar hoeveel buren moet hij kijken om het te classificeren?
```

##### 4. Train het Model (Fit)

```python
model.fit(features_train, labels_train)
```

##### 5. Maak voorspellingen (Predictions)

```python
predictions = model.predict(features_test)
```

##### 6. Beoordeel het Model

```python
score = accuracy_score(labels_test, predictions)
print(f"Model Accuracy: {score * 100}%")
```
